### Saving to Tables

```python
spark.conf.get("spark.sql.warehouse.dir")
```

```python
df1.show()
```

```python
df1.write.format("json").saveAsTable("users_json")
```

```python
spark.catalog.currentDatabase()
```

```python
spark.catalog.listTables()
```

```python
spark.sql("select * from users_json where phone is not null").show()
```

```python
df1.write \
    .format("csv") \
    .option("header", True) \
    .option("sep", "\t") \
    .saveAsTable("users_csv")
```

```python
spark.sql("select * from users_csv where phone is not null").show()
```

#### External Table

```python
df1.write \
    .format("parquet")\
    .option("path", "E:\\PySpark\\external\\users_parquet")\
    .saveAsTable("users_parquet")
```

```python
spark.catalog.listTables()
```

```python
spark.sql("select * from users_parquet where phone is not null").show()
```

```python
spark.sql("drop table users_csv")
```

```python
spark.sql("drop table users_parquet")
```
