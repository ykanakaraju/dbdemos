```python
(
  students_delta.alias("t")
    .merge(
      source = students2_df.alias("s"),
      condition = 's.student_id = t.student_id'
    )
    .whenMatchedDelete(condition = 's.action = "D"')
    .whenMatchedUpdateAll(condition = 's.action = "U"')
    .whenNotMatchedInsertAll()
    .execute()
)
```
