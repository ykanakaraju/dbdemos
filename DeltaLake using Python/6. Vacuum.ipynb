
####vacuum - Remove files no longer referenced by a Delta table####

You can remove files no longer referenced by a Delta table and are older than the retention threshold by running the ***vacuum*** command on the table. 

vacuum is not triggered automatically. 



* [Delta Table utility functions ducumentation](https://docs.delta.io/latest/delta-utility.html#remove-files-no-longer-referenced-by-a-delta-table) 

```python
%sql
USE students_db;
SELECT * FROM students;
```

```python
from delta.tables import DeltaTable

students_delta = DeltaTable.forName(spark, 'students')
```

```python
students_delta.toDF().display()
```

```python
students_delta.detail().display()
```

```python
students_delta.history().display()
```

```python
spark.read.option("versionAsOf", 5).table('students').display()
```

```python
students_delta.vacuum()  ## retention period: 168 hours (7 days)
```

```python
students_delta.vacuum(0)
```

```python
## This won't work in Serverless compute
spark.conf.set("spark.databricks.delta.retentionDurationCheck", "false")
```

```python
students_delta.vacuum(0)
```
